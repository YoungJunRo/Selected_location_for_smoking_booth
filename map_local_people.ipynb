```python
import pandas as pd

people=pd.read_csv('./data/LOCAL_PEOPLE_GU_2021.csv',encoding='euc-kr',index_col=False)
map_code=pd.read_csv('./data/Seoul_gu_code.csv',encoding='euc-kr',index_col=False)
```

```python
people_data = people.drop((['기준일ID','시간대구분']),axis=1)
```

```python
map_code_data=map_code.drop_duplicates('시군구코드')
```

```python
df=pd.merge(people_data,map_code_data,left_on='자치구코드',right_on='시군구코드')
```

```python
df_data=df.drop(['시군구코드'],axis=1)
```

```python
data=pd.pivot_table(df_data,index=['자치구코드','시군구명'],aggfunc='sum')
```
